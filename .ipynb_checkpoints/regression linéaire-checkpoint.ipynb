# Régression polynomiale

```python
%matplotlib inline
import numpy as np
import torch
import matplotlib.pyplot as plt
```

### Création des données immobilières

```python
size = 50
square_meters = np.random.randint(10, 100, size=(size))
square_meters = square_meters.reshape(size, 1)
square_meters = square_meters.astype('float32') 
```

```python
price = (square_meters ** 2) * 200 + (np.random.randn(50,1)) * 150000
price = price.astype('float32')
```

```python
plt.figure(figsize=(12,10))
plt.scatter(square_meters, price)
plt.xlabel("m²", size = 16)
plt.ylabel("Prix", size = 16)
plt.show()
```

```python
x = np.concatenate((square_meters**2, square_meters), axis=1)
print(x[:5])
print(x.shape)
```

### Conversion en Tenseur

```python
x = torch.from_numpy(x)
price = torch.from_numpy(price)
print(x.shape)
print(x.shape[1])
print(price.shape)
```

###  Création du modèle

```python
import torch.nn as nn
model = nn.Linear(x.shape[1], 1)
```

### Mise en oeuvre du modèle

```python
nb_iter = 50
lr = 1e-9
optimizer = torch.optim.SGD(model.parameters(), lr)
```

```python
loss_fn = nn.MSELoss()
```

```python
history = []
for i in range(nb_iter):
    price_pred = model(x)
    loss = loss_fn(price_pred, price)
    history.append(loss)
    if (i % 10 == 0):
        print("epoch {} : loss : {}".format(i, loss.item()))
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
```

```python
with torch.no_grad():
    plt.figure()
    plt.plot(range(nb_iter), history)
    plt.xlabel("nombre d'itération", size = 16)
    plt.ylabel("Loss", size = 16)
    plt.show()
```

### Les résultats

```python
with torch.no_grad():
    plt.figure(figsize=(12,10))
    plt.scatter(square_meters, price.numpy(), label="prix")
    plt.scatter(square_meters, price_pred.numpy(), c='red', label="prédiction")
    plt.xlabel("m²", size = 16)
    plt.ylabel("Prix", size = 16)
    plt.legend()
    plt.show()
```

```python
for name, param in model.named_parameters():
    if param.requires_grad:
        print(name, param.data)
```

```python
from sklearn.metrics import r2_score

with torch.no_grad():
    coef_determination = r2_score(price_pred.numpy(), price.numpy())
    print("Coefficient de détermination : {:.2f} %".format(coef_determination * 100))
```
